```python
world_merged = world.merge(spotify_aggregated, how='left', left_on='ISO_A3', right_on='country')
world_merged.head()
```

```python
features = ['popularity', 'danceability', 'energy', 'valence', 'tempo']
fig, ax = plt.subplots(len(features), 1, figsize=(15, 25))
fig.suptitle("Spotify Features by Country", fontsize=20)

for i, feature in enumerate(features):
    if feature not in world_merged.columns:
        print(f"Feature '{feature}' not found in merged data.")
        continue

    world_merged.plot(
        column=feature,
        cmap='YlOrRd',
        legend=True,
        ax=ax[i],
        missing_kwds={"color": "lightgrey", "label": "No data"},
        edgecolor='black',
        linewidth=0.2
    )
    ax[i].set_title(f"Average {feature.capitalize()} by Country", fontsize=14)
    ax[i].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
